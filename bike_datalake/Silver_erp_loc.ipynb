#Init

```python
import pyspark.sql.functions as F
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
RENAME_MAP = {
    "CID": "customer_id",
    "CNTRY": "country"
}
```

# Reading From Bronze

```python
df = spark.table("workspace.bronze.erp_loc")
```

# Data Transformations

## Trimming

```python
for field in df.schema.fields:
    if isinstance(field.dataType, StringType):
        df = df.withColumn(
            field.name,
            F.regexp_replace(F.trim(col(field.name)), r'-', '')
        )
```

## Renaming the Colums

```python
for old_name, new_name in RENAME_MAP.items():
  df = df.withColumnRenamed(old_name, new_name)
```

```python
df.display()
```

# Write Into Silver Table

```python
(
    df.write
      .mode("overwrite")
      .format("delta")
      .saveAsTable("workspace.silver.erp_loc")
)
```
